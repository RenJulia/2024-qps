```python
# 导入必要的库
import pandas_datareader.data as web
import matplotlib.pyplot as plt

# 获取 CPI 数据
cpi_data = web.DataReader('CPIAUCNS', 'fred')

# 计算季度通胀率
cpi_quarterly = cpi_data.resample('Q').last()  # 按季度重采样
cpi_quarterly['Inflation'] = cpi_quarterly.pct_change() * 100  # 计算通胀率

# 显示最后 4 个季度的通胀数据
cpi_quarterly.tail(4)

# 绘制 CPI 数据和通胀率图
plt.figure(figsize=(10, 6))
plt.plot(cpi_quarterly.index, cpi_quarterly['CPIAUCNS'], label='CPI')
plt.plot(cpi_quarterly.index, cpi_quarterly['Inflation'], label='Inflation Rate')
plt.legend()
plt.title('US CPI and Inflation Rate')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
